```python
def latest_check_in():
    today = pd.Timestamp.today()
    tempList = []
    for i in range(df.shape[1]):
        tempList.append(today-df['Last Check-In Date'][i])
    minValue = min(tempList)
    indexValue = tempList.index(minValue)
    
    print(df.iloc[indexValue])

latest_check_in()
```

```python
def earliest_check_in():
    today = pd.Timestamp.today()
    tempList = []
    for i in range(df.shape[1]):
        tempList.append(today-df['Last Check-In Date'][i])
    maxValue = max(tempList)
    indexValue = tempList.index(maxValue)
    
    print(df.iloc[indexValue])

earliest_check_in()
```
